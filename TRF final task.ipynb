```python
import pandas as pd
import numpy as np
from  matplotlib import pyplot as plt
%matplotlib inline
```

```python
Data =pd.read_csv("fer2013.csv")
```

```python
data=Data.values
```

```python
print(data.shape)
print(data[0, 0])
labels=data[:, 0]
```

```python
pixels=data[:,1]
print(type(pixels))
print(pixels.shape[0])
```

```python
X=np.zeros((pixels.shape[0],48*48))
print(X.shape[0])
print(X.shape)
```

```python
for ix in range(X.shape[0]):
    p=pixels[ix].split(' ')
    for iy in range(X.shape[1]):
        X[ix, iy]=int(p[iy])
```

```python
x=X
print(x)
```

```python
for ix in range (10):
    plt.figure(ix)
    plt.imshow(x[ix].reshape((48,48)),interpolation='none',cmap='gray')
plt.show()
```
